#### 04.01. 년도별 중도퇴학비율
- STD_YY : 기준년도
- TOTSUM : 총계(E)(명)
- ERSTD_CNT : 재적학생수(A)(명)
- HFWY_DROT_STDNT_RT : 중도탈락학생비율(%)((E/A) × 100)(%)

```python
# 00. 전처리 된 데이터를 load
df = pd.read_csv("./data/P02_중도탈락_학생현황_result.csv", encoding='utf8')    # 데이터

print("01. 필요한 컬럼 filter")
df = df.filter(items=['STD_YY','TOTSUM','ERSTD_CNT'])
display(df.head(2))

print("\n02. 기준년도 총계 구하기")
df = df.groupby(by='STD_YY').sum()
display(df)

print("\n03. 중도탈락 학생비율를 구하여 컬럼을 추가")
df['HFWY_DROT_STDNT_RT'] = round(df.TOTSUM/df.ERSTD_CNT*100,2)
display(df)

print("\n04. 불필요한 컬럼 삭제")
df = df.drop(columns=['TOTSUM','ERSTD_CNT'])
display(df)

print("\n05. 차트 출력")
df.index.name = '기준년도'
df.plot(kind='barh' # line, bar, barh, kde, density
      , ylim=(0,10)
      , title="년도별 중도탈락율(%)"
      , grid=True
      , legend=True
       )

""" DB 로 처리한 경우
SELECT STD_YY
     , TOTSUM
     , ERSTD_CNT
     , ROUND(TOTSUM*1.0/ERSTD_CNT*100, 2) AS HFWY_DROT_STDNT_RT
  FROM (
	  SELECT STD_YY
	       , SUM(IFNULL(TOTSUM, 0)) AS TOTSUM
	       , SUM(IFNULL(ERSTD_CNT, 0)) AS ERSTD_CNT
	    FROM API_STUDENT_OUT
	   GROUP BY STD_YY
  ) A
"""
```
