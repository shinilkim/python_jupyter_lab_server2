```python
# 파일 읽어서 dataframe 으로 변환
df = pd.read_csv("data/X02_인공지능 프로젝트 전문가_커리큘럼.csv", encoding='utf-8')
df.head(2)
```

```python
# 01. 수강 기간을 구하시오.
print("기간: ",df.일자.min()," ~ ",df.일자.max())
```

```python
# 강사별 강의 시간을 구하시오.
df.groupby(by='강사').sum().round(0)
```

```python
# 주제별 수강 시간을 구하시오.
# 정렬: 시간을 기준으로 내림차순
# 단, 수강시간이 1보다 작은 데이터는 제거
df.groupby(by='주제').sum().sort_values("시간", ascending=False).query("시간 > 0")
```
