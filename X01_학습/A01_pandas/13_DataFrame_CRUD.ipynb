#### 데이터프레임 CRUD 생성/조회/변경/삭제

```python
# 테스트 libary
import pandas as pd
import numpy as np
```

```python
# 테스트 데이터
data = {'eng' :[10,30,50,70],
        'kor' :[20,40,60,80],
        'math':[90,50,20,70]}
df = pd.DataFrame(data, index=['a','b','c','d'])
display(df)
```

```python
# 열 추가: 없으면 추가 있으면 업데이트
df['my'] = [1,2,3,4]
df
```

```python
# 열 추가: 복사
df['my1'] = df['my']
df
```

```python
# 열 추가: 연산 결과
df['my2'] = df.eng + df.kor
df
```

```python
# 행 추가
df.loc['e'] = [5,6,7,8,9,0]
df
```

```python
# 행 삭제
df.drop(index=['e'], inplace=True) # inplace: True(적용), False(미적용)
df
```

```python
# 열 삭제
df.drop(columns=['my','my1'], inplace=True)
df
```

```python
# 값 제거
df.loc['b','my2'] = np.nan
df
```

```python
# 값 제거
df.loc['b':'c', 'my2'] = np.nan
df
```

```python
# 전처리: nan 데이터 삭제
df.dropna(axis=0, inplace=False) # nan 행 삭제
```

```python
df.dropna(axis=1, inplace=False) # nan 열 삭제
```

```python
df.dropna(subset=['my2']) # my2 열에서 nan 행 삭제
```

```python
df.isnull() # nan 데이터 bool
```

```python
# nan 인 데이터 카운트
df.isnull().sum()
```

```python
# 전처리: nan 값 치환
df.fillna(999)
```

```python
# 전처리: 특정 열의 nan 값 치환
df.my2.fillna(588)
```

```python
df.sort_values(by=['my2','kor'], ascending=False)
```
