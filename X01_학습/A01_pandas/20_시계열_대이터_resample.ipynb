```python
# 테스트 libary
import pandas as pd
import numpy as np
from datetime import datetime
from pandas_datareader import data
```

```python
# 삼성전자 주식정보
sdf = data.get_data_yahoo('005930.KS', '2020-01-01')
sdf.head(2)
```

```python
# 시계열 데이터: 인덱스가 datetime 인 경우
display(sdf.loc['2020'].head(1))            # 년
display(sdf.loc['2020-02'].head(1))         # 년월
display(sdf.loc["2020-01-01":"2020-01-03"]) # 년월일 기간범위
display(sdf[sdf.index >= '2021'].head(1))   # 연산자를 이용한 기간범위
```

```python
sdf['year'] = sdf.index.year # 년도 컬럼 추가
display(sdf.head(1))
sdf['month'] = sdf.index.month # 월 컬럼 추가
display(sdf.head(1))
```

```python
g = sdf.groupby('year')
g.mean() # 년별 평균값
```

```python
sdf.groupby( sdf.index.year ).mean().filter(['High','Low','Open','Close','Volume','Adj Close']) # 년별 평균값
```

```python
sdf.groupby( [sdf.index.year, sdf.index.month] ).mean() # 년/월 평균값
```

```python
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.resample.html
sdf.resample(rule='A').mean() # 년말 집계
```

```python
sdf.resample(rule='W').mean() # 주간별 집계
```

```python
sdf.resample(rule='Q').mean() # 분기별 집계
```
