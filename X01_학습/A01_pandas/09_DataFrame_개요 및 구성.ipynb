#### 데이터 프레임(행과 열로 구성: Excel의 Sheet)
- https://pandas.pydata.org/docs/reference/frame.html
- 2차원 구조로 되어 있음

```python
import pandas as pd
import numpy as np
```

```python
d1 = [[1,2], [3,4], [5,6]]
d2 = [(1,2), (3,4), (5,6)]
d3 = [{'kor':1, 'eng':2},
      {'kor':3, 'eng':4},
      {'kor':5, 'eng':6}]
d4 = {'kor':[1,3,5], 'eng':[2,4,6]}
```

```python
df1 = pd.DataFrame(d1)
df2 = pd.DataFrame(d2)
df3 = pd.DataFrame(d3)
df4 = pd.DataFrame(d4)
display(df1)
display(df2)
display(df3)
display(df4)
```

```python
df1 = pd.DataFrame(d1, index=['a','b','c'], columns=['kor', 'eng'])
df2 = pd.DataFrame(d2, index=['a','b','c'], columns=['kor', 'eng'])
df3 = pd.DataFrame(d3, index=['a','b','c'], columns=['kor', 'eng'])
df4 = pd.DataFrame(d4, index=['a','b','c'], columns=['kor', 'eng'])
display(df1)
display(df2)
display(df3)
display(df4)
```
