```python
import matplotlib.pyplot as plt
from matplotlib.text import Text
```

```python
# 함수로 그래프를 그릴 자리를 만듭니다.
# fig는 그림을, ax는 그림에 그려질 그래프를 말합니다.
fig, ax = plt.subplots()
```

```python
# 만약 2개의 그래프를 하나의 그래프에 그리려면
fig, ax = plt.subplots(ncols=2)
```

```python
# 그래프를 세로로 그리려면 ...
fig, ax = plt.subplots(nrows=2)
```

```python
# 2행 3열 그래프
fix, ax = plt.subplots(nrows=2, ncols=3)
```

```python
# 그래프 예제
fig, ax = plt.subplots()

# (x, y) -> (0, 2), (1, 3), (2, -1)을 지나는 직선(2D) 그래프
x = [0, 1, 2]
y = [2, 3, -1]
ax.plot(x, y)
```

```python
# 2개의 그래프를 나란히 그리려면
fig, ax = plt.subplots(ncols=2)

ax[0].plot([0, 1], [2, 3]) # ax[0]은 (0, 2)와 (1, 3)을 이은 선
ax[1].plot([4, 3], [1, 2]) # ax[1]은 (4, 1)과 (3, 2)를 이은 선
```

 ### 제목, 가로축, 세로축

```python
fig, ax = plt.subplots()

ax.set_title('title')  # 그래프 제목
ax.set_xlabel('x')     # 가로축 레이블
ax.set_ylabel('y')     # 세로축 레이블

```
