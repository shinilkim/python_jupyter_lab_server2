시본(Seaborn)을 이용하여 한 연속변수의 분포를 시각화는 방법

```python
import seaborn as sns
```

데이터는 pandas의 DataFrame 형식으로 되어 있습니다.



```python
df = sns.load_dataset('tips')
df.head() # 데이터 프레임의 처음 5건을 보여 줍니다.
```

히스토그램과 밀도 추정

```python
# 커널 밀도 추정(kernel density estimation) 옵션 적용
sns.histplot(df['tip'], kde=True)
```

```python
# 커널 밀도 추정(kernel density estimation) 옵션 해제
sns.histplot(df['tip'], kde=False)
```

```python
# 커널 밀도 추정(kernel density estimation) 만을 나타내 고자 할 때
sns.kdeplot(df['tip'])
```

```python
# bins로 히스토그램에서 구간의 경계를 지정할 수 있습니다.
sns.histplot(df['tip'], bins=[0, 2, 4, 6, 8, 10])
```

```python
# color를 지정하면 그래프의 색상을 변경할 수 있습니다.
sns.histplot(df['tip'], bins=[0, 2, 4, 6, 8, 10], color='green')
```

```python
# RGB Color를 이용해 그래프의 색상을 변경할 수 있습니다.
sns.histplot(df['tip'], kde=True, color='#B53691')
```

sns.histplot(

범주형 변수의 비교

```python
import seaborn as sns
df = sns.load_dataset('tips')
df.head()
```

```python
# time 컬럼의 가짓수를 알아보고 싶다면 ... 'Dinner', 'Lunch' 두 종류가 있다는 것을 알 수 있습니다.
df['time'].unique()
```

막대 그래프

```python
# Lunch에서 팁의 평균과 Dinner에서 팁의 평균을 막대 그래프로 나타냅니다.
# 상단의 검은 막대는 95% 신뢰구간을 표시합니다.
sns.barplot(x='time', y='tip', data=df)
```

스트립플롯 :

각 범주의 평균만을 나타내면 분포에 대해 파악하기 어렵습니다.
스트립플롯은 데이터를 점으로 찍어 나타낸 것입니다.
대체로 1~4달러 정도의 범위에 팁이 집중되어 있는 것을 볼 수 있습니다.

```python
sns.stripplot(x='time', y='tip', data=df)
```

바이올린 플롯 : 바이올린 플롯은 데이터가 많은 범위는 두껍게, 데이터가 적은 범위는 얇게 그린 것입니다.

```python
sns.violinplot(x='time', y='tip', data=df)
```

# 두 연속변수의 관계

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

산점도

총액(total_bill)과 팁(tip)의 관계를 시각화해보겠습니다.
아래 그림은 점들을 뿌려놓았다고 해서 산점도(scatterplot)이라고 합니다.

```python
sns.scatterplot(x='total_bill', y='tip', data=df)
```

```python
# 색은 color로 바꿉니다.
sns.scatterplot(x='total_bill', y='tip', color='red', data=df)
```

```python
# 점의 모양(마커)을 바꿀 수도 있습니다.
sns.scatterplot(x='total_bill', y='tip', marker='+', data=df)
```

점의 불투명도는 alpha로 조정합니다.
0은 완전 투명, 1은 완전 불투명입니다.
0.2는 20%만 투명하므로 점 5개가 겹쳐야 alpha가 1일 때와 같은 정도로 불투명해집니다.

```python
sns.scatterplot(x='total_bill', y='tip', alpha=.2, data=df)
```

범주형 변수에 따라 다르게 그리기

범주형 변수에 따라 색상을 다르게 그릴 수 있습니다. hue에 범주형 변수를 지정해줍니다.



```python
sns.scatterplot(x='total_bill', y='tip', hue='time', data=df)
```

여러 개의 색상을 지정할 때는 palette를 사용합니다.

```python
sns.scatterplot(x='total_bill', y='tip', hue='time',
                palette=['purple', 'green'], data=df)
```

추세선 : 추세선을 넣을 때는 regplot을 사용합니다.
       선의 위아래로 옅게 표시된 부분은 추세선의 95% 신뢰구간을 나타냅니다.

```python
sns.regplot(x='total_bill', y='tip', data=df)
```

# 신뢰구간을 지우려면 ci=None을 해줍니다.

```python
sns.regplot(x='total_bill', y='tip', ci=None, data=df)
```

# lowess=True를 하면 추세선을 선형이 아닌 국소 회귀 기법으로 그립니다.
# 이 기법은 부분 부분의 선형 추세를 찾은 다음에 그 추세들을 이어 매끄러운 곡선으로 만니다.

```python
sns.regplot(x='total_bill', y='tip', lowess=True, data=df,
            line_kws={'color': 'red'}, scatter_kws={'color': 'grey'})
```

그리드 : 그리드를 이용하면 범주형 변수로 그래프를 나눠서 그릴 수 있습니다.

```python
import seaborn as sns
df = sns.load_dataset('tips')
df.head()
```

다음은 time에 따라 그래프를 행으로 나눠 그리고,
각 그래프는 가로축은 total_bill, 세로축은 tip인 산점도(scatterplot)을 그리는 예입니다.

```python
grid = sns.FacetGrid(data=df, row='time')
grid.map(sns.scatterplot, "total_bill", "tip")
```

다음은 day에 따라 그래프를 열로 나눠 그리고,
각 그래프는 가로축은 time, 세로축은 tip인 stripplot을 그리는 예입니다.



```python
grid = sns.FacetGrid(data=df, col='day')
grid.map(sns.stripplot, 'time', 'tip')
```

row_order로 행의 순서를, col_order로 열의 순서를 지정할 수 있습니다.

```python
grid = sns.FacetGrid(data=df, col='smoker', col_order=['No', 'Yes'])
grid.map(sns.scatterplot, "total_bill", "tip")
```
